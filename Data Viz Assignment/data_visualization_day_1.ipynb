######1. load the iris.csv dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('iris.csv')
data.head()
```

#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
p=sns.countplot(data=data,x='Species')
p.bar_label(p.containers[0],size=13)
sns.despine()
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here
data['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%')
plt.title("Iris Species %")
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=data, color='blue') 
plt.show()
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
sns.scatterplot(data=data, x='SepalLengthCm',y='PetalWidthCm', hue='Species')
plt.show()
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
### code here
sns.boxplot(y=data['SepalLengthCm'])
plt.show()
```

```python
sns.boxplot(x='Species',y='SepalLengthCm',data=data,linewidth=2)
plt.show()
```

###### 6. Create a Strip plot 

```python
### code here
sns.stripplot(x='Species',y='SepalLengthCm',data=data)
plt.show()
```

###### 7. Create a Violin Plot.

```python
### code here
sns.violinplot(x='Species',y='SepalLengthCm',data=data)
plt.show()
```

##### 8. Create a pair chart.

```python
### code here
sns.pairplot(data=data)
plt.show()
```

##### 9. Create a Heat Map

```python
### code here
sns.heatmap(data.corr(),annot=True)
plt.show()
```

###### 10. Create a distribution plot for all numerical colomns.

```python
### code here
plt.figure(figsize=(10, 5))
sns.set_style('white')
plt.subplot(2,2,1)
sns.kdeplot(data['SepalLengthCm'], color ='blue')
plt.subplot(2,2,2)
sns.kdeplot(data['SepalWidthCm'], color ='blue')
plt.subplot(2,2,3)
sns.kdeplot(data['PetalLengthCm'], color ='green')
plt.subplot(2,2,4)
sns.kdeplot(data['PetalWidthCm'], color ='green')
plt.show()
```
