##### import the titanic dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

##### 1. Write a Python program to display a bar chart of the fare of class.

```python
### code here
sns.barplot(x='class',y='fare',data=df,color='b')
plt.show()
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
### code here
sns.barplot(y='class',x='fare',data=df,color='r')
plt.show()
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
### code here
sns.barplot(x='class',y='fare',data=df,palette='afmhot')
plt.show()
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
### code here
df.groupby(['class']).sum().plot(kind='pie', y='fare',autopct='%1.0f%%')
plt.show()
```

###### 5. Write a Python program to display a count plot of all categorical columns.

```python
### code here
plt.figure(figsize=(18,10))
plt.subplot(2,3,1)
sns.countplot(x=df.sex)
plt.subplot(2,3,2)
sns.countplot(x=df.embarked)
plt.subplot(2,3,3)
sns.countplot(x='class',data=df)
plt.subplot(2,3,4)
sns.countplot(x=df.who)
plt.subplot(2,3,5)
sns.countplot(x=df.alive)
plt.subplot(2,3,6)
sns.countplot(x=df.alone)
plt.show()
```

###### 6. Write a Python program to display a factor plot of ( pclass vs embarked). hist (kind = 'count')

```python
  #### code here
sns.catplot(x ='embarked',kind ='count', col ='pclass', data = df)
plt.show()
```

###### 7. Write a Python programming to display a factor plot of ( pclass vs sex). hist (kind = 'count')

```python
### code here
sns.catplot(x ='sex',kind ='count', col ='pclass', data = df)
plt.show()
```

###### 8. Write a Python programming to display a factor plot of ( embarked vs sex). hist (kind = 'count')

```python
### code here 
sns.catplot(x ='embarked',kind ='count', col ='sex', data = df)
plt.show()
```

###### 9. Write a Python program to display distribution plot for all numberical columns.

```python
### code here
plt.figure(figsize=(15, 10))
sns.set_style('white')
plt.subplot(2,2,1)
sns.kdeplot(df['age'], color ='blue')
plt.subplot(2,2,2)
sns.kdeplot(df['fare'], color ='blue')
plt.subplot(2,2,3)
sns.kdeplot(df['survived'], color ='green')

```

###### 10. Write a Python program to display boxplot of age, pclass and hue = gender.

```python
### code here
sns.boxplot(x='pclass',y='age',data=df,hue='sex')
plt.show()
```
