```python
import numpy as np
x=['Python','C','C++','Jave']
y=[85,70,60,82]
z=[25,50,35,68]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Language',fontsize=15)   
plt.ylabel('Percentage',fontsize=15)
plt.title('Popularity',fontsize=15)
  
plt.bar(p,y,width,color='r',label='popularity')
plt.bar(p1,z,width,color='b',label='popularity1')

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
x=['Python','C','C++','Jave']
y=[85,70,60,82]
z=[25,50,35,68]

plt.xlabel('Language',fontsize=15)   
plt.ylabel('Percentage',fontsize=15)
plt.title('Popularity',fontsize=15)
  
plt.barh(x,y,color='r',label='popularity')        #horizontal bar graph
plt.barh(x,z,color='b',label='popularity1')     
plt.legend()
plt.show()
```

## Fill_Between Plot

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.show()
```

```python
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color='r')
plt.fill_between(x,y,where=(x>=2) & (x<=4),alpha=0.3)
plt.show()
```
