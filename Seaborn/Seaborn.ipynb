## HeatMap

```python
import numpy as np
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=10)
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=10,cmap='gist_heat')
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=10,annot=True)
plt.show()
```

```python
y={'fontsize':15,'color':'b'}
sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y)
plt.show()
```

```python
v=sns.heatmap(var,vmin=0,vmax=10,annot=True)
v.set(xlabel='python',ylabel='java')
sns.set(font_scale=1)
plt.show()
```
